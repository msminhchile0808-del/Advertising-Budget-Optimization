Create an optimized solution to allocate a budget across three advertising channels: Online Ads, TV Ads, and Print Ads. Each channel has different costs, reach, and impact. In addition, we must pay for at least one unit of each of the three channels or else we lose our contract with the distributors. Use the parameters and contraints identified to create a solution.

**Parameters**
* Total budget: $10,000

**Channel costs, reach, and impact per unit**
* Online ads:
  * *Cost*: 50 USD per unit
  * *Reach*: 1,000 people per unit
  * *Impact*: 2 (this is a relative measure of impact)
* TV ads:
  * *Cost*: 200 USD per unit
  * *Reach*: 5,000 people per unit
  * *Impact*: 5
* Print ads:
  * *Cost*: 100 USD per unit
  * *Reach*: 3,000 people per unit
  * *Impact*: 3

```python
# Print the optimal number of units for each channel and the overall reach * impact for the solution

!pip install pulp
import pulp

# Total budget
budget = 10000

# Online ads:
Online_ads_cost = 50      #USD per unit
Online_ads_Reach = 1000  #people per unit
Online_ads_Impact = 2     #(this is a relative measure of impact)
#TV ads:
TV_ads_Cost = 200         #USD per unit
TV_ads_Reach = 5000      #people per unit
TV_ads_Impact = 5
#Print ads:
Print_ads_Cost = 100      # USD per unit
Print_ads_Reach = 3000   #people per unit
Print_ads_Impact = 3

# Create the LP problem
prob = pulp.LpProblem("Advertising_Budget_Allocation", pulp.LpMaximize)

# Define decision variables for the number of units for each channel
# The upper bound is set to a large enough number to not constrain the problem,
# as the budget constraint will be the limiting factor.
online_units = pulp.LpVariable("Online Units", lowBound=1, cat='Integer')
tv_units = pulp.LpVariable("TV Units", lowBound=1, cat='Integer')
print_units = pulp.LpVariable("Print Units", lowBound=1, cat='Integer')

# Define the objective function (Maximize total reach * impact)
prob += (online_units * Online_ads_Reach * Online_ads_Impact +
         tv_units * TV_ads_Reach * TV_ads_Impact +
         print_units * Print_ads_Reach * Print_ads_Impact), "Total Reach x Impact"

# Define the budget constraint
prob += (online_units * Online_ads_cost +
         tv_units * TV_ads_Cost +
         print_units * Print_ads_Cost) <= budget, "Budget Constraint"


# Solve the problem
prob.solve()

# Print the results
print(f"Online Ads Units: {online_units.varValue} units")
print(f"TV Ads Units: {tv_units.varValue} units")
print(f"Print Ads Units: {print_units.varValue} units")

# Calculate and print the overall reach * impact
optimal_reach_impact = (online_units.varValue * Online_ads_Reach * Online_ads_Impact +
                        tv_units.varValue * TV_ads_Reach * TV_ads_Impact +
                        print_units.varValue * Print_ads_Reach * Print_ads_Impact)
print(f"Overall Reach x Impact: {optimal_reach_impact}")
```
